```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("../data/great_zimbabwe.png")
```

```python
print(image.shape)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("original image")
plt.axis("off")
```

## Raw Pixel Values

```python
print(image_rgb)
```

## Grayscale Image

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.title("Gray scale image")
plt.axis("off")
```

```python
print(gray.shape)
```

```python
print(gray)
```

## Resize image

```python
resized_img = cv2.resize(image_rgb, (300, 300))
plt.imshow(resized_img)
plt.title("Resized")
plt.axis("off")
```

## Flip Image

```python
flipped_img = cv2.flip(image_rgb,0)
plt.imshow(flipped_img)
plt.title("flipped image")
plt.axis("off")
```

## Brightness

```python
brighter = cv2.convertScaleAbs(image_rgb, alpha=1, beta=20)

plt.imshow(brighter)
plt.title("Brighter")
plt.axis("off")
```
